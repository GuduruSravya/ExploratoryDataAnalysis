## Find out the total number of doctorates who are married

```python

# Find out the total number of doctorate who are married

num_doctorate_civ_spouse = len(df_data[df_data['education'].str.match('Doctorate')& (df_data['marital-status'].str.match('Married-civ-spouse'))])
num_doctorate_spouse_absent =  len(df_data[df_data['education'].str.match('Doctorate')& (df_data['marital-status'].str.match('Married-AF-spouse'))])
num_doctorate_AF_spouse = len(df_data[df_data['education'].str.match('Doctorate')& (df_data['marital-status'].str.match('Married-spouse-absent'))])

num_doctorate = num_doctorate_civ_spouse+num_doctorate_spouse_absent+num_doctorate_AF_spouse
print ("\n\nThe total number of doctorates who are married %s"
        % (num_doctorate))

```

# Visualization of data


### Plot the race count for each country

```python
sns.set_style('whitegrid')
sns.set(font_scale = 1.3)


unique_country = df_data['native-country'].unique().tolist()

race_data=[]
for country in unique_country:
    country_data = df_data[df_data['native-country'].str.contains(country,regex=False)]
    race_count = len(country_data['race'].unique().tolist())
    race_data.append(race_count)

sns.set(rc={'figure.figsize':(25,6)})
ax=sns.barplot(x=unique_country,y=race_data)
ax.set_xlabel("Native-Country",size=24)
ax.set_ylabel("RaceCount",size=24)
ax.set_title("Race count for each country",size=30)

labels=ax.get_xticklabels()
dummy=ax.set_xticklabels(labels, rotation=90,fontsize=18)





```

### Pie chart that represents native country

```python

native_country = df_data['native-country'].unique().tolist()
country_count = df_data['native-country'].value_counts()
data = country_count.values.tolist()

x=np.char.array(native_country)
y=np.array(data)

porcent = 100.*y/y.sum()

patches, texts = plt.pie(y,radius=4)
labels = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(x, porcent)]


plt.legend(patches, labels, loc='center right', bbox_to_anchor=(4,0.5),
           fontsize=14)

plt.show()
```
